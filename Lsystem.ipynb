```python
import re
# import turtle
from turtle import *
```

```python
class Lsystem():
    def __init__(self):
        self.strings = "F"
        self.rule1 = ["F","FF+[+F-F-F]-[-F+F+F]"]
        #self.rule1[["A","B"],["B","AB"]]
        self.distance = 10
        self.theta = 30
        self.t = Turtle()
        self.t.screen.bgcolor("black")
        self.t.pencolor("white")
#         self.t.pencolor(200,0,0,0)
        self.t.screen.setup(1000,800)
        self.t.lt(90)
        self.t.penup()
        self.t.goto(0,-300)
        self.t.pendown()
        self.t_pos = []
        self.t_ang = []
        
    def replacement(self,strings):
        strings = re.sub(self.rule1[0],self.rule1[1],strings)
        return strings
    def draw(self,strings):
        list_str = list(strings)
        self.t.speed(10)
        for s in list_str:
            if s == "F":
                self.t.fd(self.distance)
            elif s == "+":
                self.t.lt(self.theta)
            elif s == "-":
                self.t.rt(self.theta)
            elif s == "[":
                self.t_pos.append(self.t.position())
                self.t_ang.append(self.t.heading())
            elif s == "]":
                self.t.penup()
                self.t.goto(self.t_pos[-1])
                self.t.setheading(self.t_ang[-1])
                self.t.pendown()
                self.t_pos.pop(-1)
                self.t_ang.pop(-1)
                #self.distance = self.distance -2
```

```python
l = Lsystem()
for i in range(30):
    if i == 0:
        print("init:",l.strings)
        strings = l.replacement(l.strings)
    else:
        strings = l.replacement(strings)
#     l.draw(strings)
    print("{0}:".format(i),strings)
    print("-----------------------------------------------------------")
#l.draw(strings)
#exitonclick()
print("Finish")
```
