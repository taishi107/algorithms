```python
class Lsystem():
    def __init__(self):
        self.rules = {
            "!(1)":"!(3)",
            "F(s)":"G(s)[-!(1)F(s)]",
        }
    def replacement(self,strings):
        ans = []
        #t = list(mojiretu)
        for s in strings:
            ans.append(self.rules[s])
        strings = ''.join(ans)
        return strings
```

```python
strings = "A"
```

```python
l = Lsystem()
strings = l.replacement(strings)
print(strings)
```
